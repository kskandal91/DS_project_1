```python
# NumPy basics: creation, dtype, shape, indexing
import numpy as np
# Create arrays
a = np.array([1, 2, 3])
b = np.arange(6).reshape(2, 3)
c = np.linspace(0, 1, 5)
print('a:', a)
print('b:\n', b)
print('c:', c)
print('dtypes:', a.dtype, b.dtype, c.dtype)
print('shapes:', a.shape, b.shape, c.shape)
# Indexing and slicing
print('b[0,1]:', b[0, 1])
print('b[:,1]:', b[:, 1])
```

```python
# Operations & broadcasting
x = np.array([10, 20, 30])
y = np.array([[1], [2], [3]])  # shape (3,1)
print('x + 5 ->', x + 5)
print('y * 2 ->\n', y * 2)
# Broadcasting: (3,1) + (3,) -> (3,3)
print('y + x ->\n', y + x)
# Aggregation
m = np.arange(12).reshape(3,4)
print('m:\n', m)
print('m.sum(axis=0):', m.sum(axis=0))
print('m.mean():', m.mean())
```

```python
# Linear algebra examples
A = np.array([[1.0, 2.0], [3.0, 4.0]])
v = np.array([1.0, 1.0])
print('A:\n', A)
print('v:', v)
print('A @ v:', A @ v)
print('np.dot(A, v):', np.dot(A, v))
print('inv(A):\n', np.linalg.inv(A))
print('eigvals(A):', np.linalg.eigvals(A))
```

```python
import numpy as np

def pca_numpy(X, n_components=2):
    """
    Perform PCA using NumPy only.
    
    Parameters
    ----------
    X : ndarray of shape (n_samples, n_features)
        The data matrix.
    n_components : int
        Number of principal components to return.
        
    Returns
    -------
    X_pca : ndarray of shape (n_samples, n_components)
        Projected data
    components : ndarray of shape (n_components, n_features)
        Principal component directions (eigenvectors)
    explained_variance : ndarray of shape (n_components,)
        Eigenvalues associated with the selected components
    """

    # 1. Center the data (subtract mean)
    X_meaned = X - X.mean(axis=0)

    # 2. Compute covariance matrix
    cov_matrix = np.cov(X_meaned, rowvar=False)

    # 3. Compute eigenvalues & eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

    # 4. Sort eigenvalues (and corresponding eigenvectors) in descending order
    sorted_idx = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_idx]
    eigenvectors = eigenvectors[:, sorted_idx]

    # 5. Select top n_components
    components = eigenvectors[:, :n_components]
    explained_variance = eigenvalues[:n_components]

    # 6. Project data onto principal components
    X_pca = np.dot(X_meaned, components)

    return X_pca, components, explained_variance


# ----------------- Test Example -----------------

# create a toy dataset
np.random.seed(0)
X = np.random.rand(10, 3)  # 10 samples, 3 features

# compute PCA
X_pca, components, explained_var = pca_numpy(X, n_components=2)

print("Projected data:\n", X_pca)
print("\nPrincipal components (eigenvectors):\n", components)
print("\nExplained variance (eigenvalues):\n", explained_var)
```
